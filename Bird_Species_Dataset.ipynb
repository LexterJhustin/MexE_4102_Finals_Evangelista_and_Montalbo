# Revised Topic: Converting images of Six Bird Species into Grayscale

#### American Goldfinch

```python
import cv2  #cv2 openCV(Open Source Computer Vision Library). used for image processing.
from google.colab.patches import cv2_imshow #used for displaying images in google colab

# Colorful image - 3 Channels
img = cv2.imread("drive/MyDrive/Bird Species Dataset/AMERICAN GOLDFINCH/001.jpg") #used for reading images
print(img.shape) #used for printing the shape of the image
cv2_imshow(img) #used for displaying the image

#Grayscale image
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #used for converting the image to grayscale
print(gray.shape)
cv2_imshow(gray)
```
